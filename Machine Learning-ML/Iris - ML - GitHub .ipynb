```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
#Diviser les données en ensemble d'entrainement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Normaliser les caractéristiques

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() #répartition standard dans une courbe
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Initialiser les modèles

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

models = {
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "Random Forest": RandomForestClassifier(),
    "Logistic Regression": LogisticRegression()
}
```

```python
#Entrainer et évaluer chaque modèle

from sklearn.metrics import classification_report, confusion_matrix

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Model: {name}")
    print(confusion_matrix(y_test, y_pred)) #confusion matrix c'est pour savoir après l'apprentissage, quand il a fait une prédiction, de vérifier son taux d'erreur.
    print(classification_report(y_test, y_pred))
    print("\n")
    
#0, 1 et 2 sont les classes de fleur.
```

```python
#Visualisation
#Réduction de dimensionnalité avec PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Visualisation des données en 2D

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=iris.target_names[y], palette="viridis")
plt.title("Projection 2D des données Iris après PCA")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```
